```python
import pandas as pd
import numpy as np

# Crear un dataframe con datos de 2022 a 2024
np.random.seed(42)  # Para reproducibilidad

# Número de registros  por año
records_per_year = 1000
years = [2022, 2023, 2024]
data = []

# Valores posibles para algunas columnas basados en el análisis original
departamentos = ['Antioquia', 'Boyacá', 'Córdoba', 'Cundinamarca', 'Santander']
condiciones_meteor = ['Soleado', 'Lluvia', 'Nublado']
estado_super = ['Seca', 'Húmeda']
terrenos = ['Montañoso', 'Plano']
geometria_acc = ['Curva', 'Recta']
clases_accidente = ['Colisión', 'Volcamiento', 'Atropello']
causas_posibles = ['No mantener distancia', 'Exceso de velocidad', 'Inexperiencia']

for year in years:
    for i in range(records_per_year):
        record = {
            'AÑO': year,
            'DEPARTAMENTO': np.random.choice(departamentos),
            'FECHA_ACC': f"{year}-{np.random.randint(1,13):02}-{np.random.randint(1,29):02}",
            'DIA_SEMANA_ACC': np.random.choice(['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']),
            'CONDIC_METEOR': np.random.choice(condiciones_meteor),
            'ESTADO_SUPER': np.random.choice(estado_super),
            'TERRENO': np.random.choice(terrenos),
            'GEOMETRIA_ACC': np.random.choice(geometria_acc),
            'N_HERIDOS': np.random.poisson(0.6),
            'N_MUERTOS': np.random.binomial(1, 0.1),
            'N_VICTIMAS': lambda x: x['N_HERIDOS'] + x['N_MUERTOS'],
            'CLASE_ACCIDENTE': np.random.choice(clases_accidente),
            'CAUSA_POSIBLE': np.random.choice(causas_posibles)
        }
        record['N_VICTIMAS'] = record['N_HERIDOS'] + record['N_MUERTOS']
        data.append(record)

# Convertir a DataFrame
df = pd.DataFrame(data)
df.head()
```
